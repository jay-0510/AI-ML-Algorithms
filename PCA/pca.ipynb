# Principal Component Analysis - Practical Implementation

### PCA - Used for Feature transformation in the ML. 
### Here, from IRIS DATASET - We've 4 features & we'll reduce into 2 features.
### 4D --> 2D

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()
X = pd.DataFrame(iris.data)
y = iris.target
```

```python
X.head()
```

```python
# Standarise the Data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled.shape
```

### Apply PCA - Linear dimensionality Reduction

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
X_pca.shape
```

```python
# Checkout the variance inside the Model 
print("Explained Variance Ratio: ", pca.explained_variance_ratio_)
```

```python
print(pca.components_) # Gives EIGEN VECTORS
```

```python
# Visualise the IRIS data of PCA
plt.figure(figsize = (8, 6))

plt.scatter(
    X_pca[:, 0],
    X_pca[:, 1],
    c = y
)

plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.title("PCA with IRIS Dataset")
```

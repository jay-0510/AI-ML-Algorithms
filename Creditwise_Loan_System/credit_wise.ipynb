## Handling Missing Values

```python
# Categorical & Numerical data — Both have different ways to handle 
categorical_cols = df.select_dtypes(include=["object"]).columns
numerical_cols = df.select_dtypes(include=["float64"]).columns
```

### Sklearn imputer -- completing the missing values with missing strategies
#### Replace the missing values with the mean, median, most frequent, constant values

```python
from sklearn.impute import SimpleImputer
num_imp = SimpleImputer(strategy="mean")
df[numerical_cols] = num_imp.fit_transform(df[numerical_cols])
```

```python
df.head()
```

```python
cat_imp = SimpleImputer(strategy="most_frequent")
df[categorical_cols] = cat_imp.fit_transform(df[categorical_cols])
```

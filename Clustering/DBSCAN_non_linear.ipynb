# DBSCAN Clustering

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# Scale the data 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Visualsation of data 
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:,2])
```

```python
# Implement DBSCAN 

from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps = 0.7, min_samples = 5)

labels = dbscan.fit_predict(X_scaled)
```

```python
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:,2], c = labels)
```

# We'll apply K means and DBSCAN, So we can understand difference between them.

```python
# Kmeans -- To apply this data should be divided into 2 Cluster.
from sklearn.cluster import KMeans

kmeans = KMeans(
    n_clusters=2,
    random_state=42
)

labels = kmeans.fit_predict(X_scaled)
```

```python
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:,1], c = labels)
```

```python
# DBSCAN 
dbscan = DBSCAN(
    eps=0.5,
    min_samples=5
)
dbscan.fit_predict(X_scaled)
```

```python
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:,1], c = labels)
```
